```python
# 1. Start with daily zonal mean v anomaly (filtered)
v_zm_anom = highpVshaved  # dims: (alltime, level, latitude)

# 2. Calculate meridional convergence: -∂v̄/∂y
R_earth = 6.371e6  # meters
v_merid_conv = -v_zm_anom.differentiate('latitude') * (180/np.pi) / R_earth

# 3. Select equator and cross-correlate with Z(850)
v_conv_corr = uf.level_lag_corr(highpZshaved.sel(level=850, latitude=0), v_merid_conv.sel(latitude=0))

# 4. Calculate std of the convergence at equator
v_conv_std = v_merid_conv.sel(latitude=0).std(dim='alltime')

# 5. Convert correlation to covariance
v_conv_cov = v_conv_corr * v_conv_std  # Units: [1/s]

# 6. Convert to [1/day]
v_conv_cov_per_day = v_conv_cov * 86400  # [1/d]
```

```python
plt.figure(figsize=(20, 5))
cs = plt.contourf(
    v_conv_cov_per_day.lag, v_conv_cov_per_day.level, v_conv_cov_per_day.values,
    levels=21, cmap="RdBu_r", extend="both",
    norm=TwoSlopeNorm(vmin=v_conv_cov_per_day.min(), vcenter=0, vmax=v_conv_cov_per_day.max())
)
plt.gca().invert_yaxis()
plt.colorbar(cs, label="[1/day]")
plt.yscale('log')
plt.gca().set_yticks(pressure_levels)
plt.gca().set_yticklabels([str(int(p)) for p in pressure_levels])
plt.gca().set_ylim(1000, 100)
plt.gca().yaxis.set_major_locator(LogLocator(base=10.0, subs=[], numticks=5))
plt.gca().set_yticks(pressure_levels)
plt.gca().set_yticklabels([str(int(p)) for p in pressure_levels])
plt.xlabel("Lag (days)")
plt.ylabel("Pressure (hPa)")
plt.title(r"$-(\bar{v})_y$ Variability at the Equator (1y Highpass Filtered)")
plt.savefig('meridconvxz850eqcov.png', dpi=300, bbox_inches='tight') 
plt.show()
```
